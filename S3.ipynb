## Dictionaries

```python
dict1 = {
    "01": "Astana",
    "02": "Almaty",
    "03": "Aqmola region",
    "04": "Aqtobe region",
    "05": "Almaty region",
    "06": "Atyrau region",
    "07": "West Qazaqstan region",
    "08": "Zhambyl region",
    "09": "Qaragandy region",
    "10": "Qostanay region",
    "11": "Qyzylorda region",
    "12": "Mangystau region",
    "13": "Turkestan region",
    "14": "Pavlodar region",
    "15": "North Qazaqstan region",
    "16": "East Qazaqstan region",
    "17": "Shymkent region",
    "18": "Abai region",
    "19": "Zhetysu region",
    "20": "Ulytau region"
}
```

```python
plate_numbers = ["123ABC02", "237ACD04", "357DDO14"]

result = dict()

for number in plate_numbers:
    key = number[-2:]
    if key in dict1.keys():
        result[number] = dict1[key]

for key, value in result.items():
    print(f"Car number {key} belongs to {value}")
```
