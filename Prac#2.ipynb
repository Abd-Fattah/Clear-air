## Задача:

1. Для массива pm найти среднее арифметическое и медиану, и отобразить их на графике
2. Для массива температур найти среднее арифметическое и медиану, и отобразить их на графике

*Среднее арифметическое* - это сумма всех значений, поделенная на количество значений

*Медиана* - это значение среднего по величине элемента


```python
sum_of_pm = sum(pm)

avg_pm = sum_of_pm/len(pm) 

avg_list_pm = []

avg_list_pm.append(avg_pm)

med_list_pm = []

sorted_pm = sorted(pm)

for i in pm:
    med_list_pm.append(pm[25])
```

```python
plt.plot(x, pm, linewidth=2.0, color="b", label="PM2.5")
plt.plot(x, avg_list_pm*50, linewidth=2.0, color="r", label="Avg_pm")
plt.plot(x, med_list_pm, linewidth=2.0, color="g", label="Med_pm")

plt.legend() 
plt.show()
```
