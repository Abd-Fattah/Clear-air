```python
import random

rand_list = []
n = 10

for i in range(0, n+1):
    rand_list.append(random.randint(1, 51))
    

# Python program for implementation of Bubble Sort

def bubbleSort(arr):
	n = len(rand_list)
	swapped = False
	for i in range(n-1):

		for j in range(0, n-i-1):
			if arr[j] > arr[j + 1]:
				swapped = True
				arr[j], arr[j + 1] = arr[j + 1], arr[j]
		
		if not swapped:
			return

print("Not sorted array is:", end=" ")        
        
for i in range(len(rand_list)):
	print("% d" % rand_list[i], end=" ")
                
bubbleSort(rand_list)

print("")

print("Sorted array is:", end=" ")
for i in range(len(rand_list)):
	print("% d" % rand_list[i], end=" ")
```

```python
import random
import matplotlib.pyplot as plt

rand_list = []
n = 512

for i in range(0, n+1):
    rand_list.append(random.randint(1, 51))
    

def bubbleSort(arr):
    n = len(rand_list)
    swapped = False
    for i in range(n-1):

        for j in range(0, n-i-1):
            if arr[j] > arr[j + 1]:
                swapped = True
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
        if not swapped:
            return

print("Not sorted array is:", end=" ")        
        
for i in range(len(rand_list)):
	print("% d" % rand_list[i], end=" ")
    
fig, ax = plt.subplots()
ax.plot(rand_list)
plt.show()
                
bubbleSort(rand_list)

print("")

print("Sorted array is:", end=" ")
for i in range(len(rand_list)):
	print("% d" % rand_list[i], end=" ")
    
fig, ax = plt.subplots()
ax.plot(rand_list)
plt.show()
```
